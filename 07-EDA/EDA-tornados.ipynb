````python
url = "https://github.com/esnt/Data/raw/main/CleanData/tornadoes.csv"
df = pd.read_csv(url)
df["lon"]= df['lon']*-1
````

````python
df.head()
````

````python
fig = px.scatter(df, x = "lon", y = "lat", hover_data=df.columns, color = 'state')
fig.update_layout(width=800, height=600)
````

````python
px.scatter_geo(df, lon = 'lon', lat='lat', scope = 'usa')
````

````python
px.box(df, x  ='max_width')
````

````python
counts = df['state'].value_counts()
counts = counts[~counts.index.str.contains(',')]
````

````python
fig = px.choropleth(counts, locations = counts.index, locationmode="USA-states",
              scope = 'usa', color = 'count')
fig.show()
````
